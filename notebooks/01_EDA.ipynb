# Heart Disease — Exploratory Data Analysis

**Dataset**: UCI Heart Disease (Cleveland). 303 patients, 13 features, binary target (`target` = 1 indicates presence of disease).

This notebook accompanies the report. Re-run it after `python -m src.data.download_data`.

```python
import sys, pathlib
sys.path.append(str(pathlib.Path.cwd().parent))
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')
df = pd.read_csv('../data/processed/heart_disease.csv')
print(df.shape)
df.head()
```

```python
df.info()
df.describe().T
```

## Class balance

```python
df['target'].value_counts().rename({0:'no_disease', 1:'disease'}).plot.bar(rot=0)
plt.title('Class balance'); plt.ylabel('count'); plt.show()
```

## Missing values

```python
df.isna().sum().sort_values(ascending=False).head(10)
```

## Numeric distributions by target

```python
import numpy as np
fig, axes = plt.subplots(2, 3, figsize=(13,7))
for ax, col in zip(axes.flat, ['age','trestbps','chol','thalach','oldpeak']):
    sns.histplot(data=df, x=col, hue='target', kde=True, bins=25, ax=ax); ax.set_title(col)
axes.flat[-1].set_visible(False)
plt.tight_layout(); plt.show()
```

## Correlation heatmap

```python
plt.figure(figsize=(9,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature correlation'); plt.show()
```

## Categorical features vs target

```python
fig, axes = plt.subplots(2,4, figsize=(15,7))
for ax, col in zip(axes.flat, ['sex','cp','fbs','restecg','exang','slope','thal']):
    sns.countplot(data=df, x=col, hue='target', ax=ax); ax.set_title(col)
axes.flat[-1].set_visible(False)
plt.tight_layout(); plt.show()
```

## Key takeaways
* Dataset is roughly balanced (≈ 54% / 46%), so accuracy is a usable headline metric, but ROC-AUC is reported alongside.
* `ca` and `thal` contain a few missing values — handled via median / most-frequent imputation in the pipeline.
* `cp` (chest-pain type), `thalach` (max heart rate) and `oldpeak` show the strongest visual separation across classes.
